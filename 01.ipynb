```python
# Import lớp Image
from PIL import Image
```

```python
# Thư mục chứa hình ảnh
dir = 'C:/Users/HP/image'
```

```python
# Đường dẫn đến hình ảnh
image_path = dir +'/1.JPG'
print(image_path)
```

```python
# Sử dụng Image.open() đọc ảnh
img01 = Image.open(image_path)
```

```python
# Hiển thị hình ảnh
display(img01)
```

```python
# In một số thông tin
# In định dạng ảnh
print("Định dang ảnh: ", img01.format)
```

```python
# In ra size
print("Kích thước hình ảnh: ", img01.size)
```

```python
new_img_01_path = 'C:/Users/HP/image/1.jpg'
new_img_png_01_path = 'C:/Users/HP/image/1.png'
img01.save(new_img_01_path)
img01.save(new_img_png_01_path)
```
